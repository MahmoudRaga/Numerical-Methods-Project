```python
#some imports
import sys
from PyQt5 import (  QtGui, QtCore  )
from PyQt5.QtWidgets import (  QApplication, QLabel, QLineEdit, QPushButton, QWidget, QGridLayout  )
from PyQt5.QtGui import (  QPixmap, QCursor, QIcon  )
from PyQt5.QtCore import Qt
from math import *
from scipy import misc

#global dictionary to store the widgets
widgets = {
    "logo": [],
    "button-one": [],
    "button-two": [],
    "button-three": [],
    "button-four": [],
    "button-fife": [],
    "button-six": [],
    "label-one": [],
    "label-two": [],
    "label-three": [],
    "label-four": [],
    "input-one": [],
    "input-two": [],
    "input-three": [],
    "input-four": [],
    "output": []
}

#creating a window and styling it
app = QApplication(sys.argv)
window = QWidget()
window.setWindowTitle("Numerical Methods")
window.setWindowIcon(QtGui.QIcon('./imgs/icon.ico'))
window.setFixedSize(1000, 500)
window.move(200, 100)
window.setStyleSheet(
    "background: #17223B;"
    )

#setting the window layouts
mainLayout = QGridLayout()
grid = QGridLayout()

#deleting all widgets in the window
def clearWidgets():
    #hide all existing widgets and erase them from the global dictionary
    for widget in widgets:
        if widgets[widget] != []: # if its empty 
            widgets[widget][-1].hide() # (-1) because any list or dictionary or tuple start with (0)
        for i in range(0, len(widgets[widget])):
            widgets[widget].pop() # (pop()) for remove it from the dictionary because of (hide()) make them hidden only

#===============================MOVING BETWEEN FRAMES====================================
#starting the program by clicking on the start button
def start():
    #calling the clear function
    clearWidgets()
    #calling another function to set another frame
    choicesFrame()

#back to the 1st frame
def backToStart():
    #calling the clear function
    clearWidgets()
    #calling another function to set another frame
    startingFrame()

#back to choices frame
def backToChoices():
    #calling the clear function
    clearWidgets()
    #calling another function to set another frame
    choicesFrame()
#==================================================================================

#===============================UTILITIES====================================
#creating a logo by one function
def createLogo(address):
    logo = QLabel()
    logo.setPixmap(QPixmap(address))
    logo.setAlignment(QtCore.Qt.AlignCenter)
    widgets["logo"].append(logo)
    mainLayout.addWidget(logo, 0, 0, alignment=Qt.AlignCenter)
    return logo

#creating a label by one function
def createLabel(label_text, memory_name, row, column):
    label = QLabel(label_text)
    label.setStyleSheet(
        "color: #fff;"+
        "font-family: 'Rounded Mplus 1c';"+
        "font-size: 16px;"+
        "margin: 5px 0 5px 100px;"
    )
    widgets[memory_name].append(label)
    grid.addWidget(label, row, column, alignment=Qt.AlignLeft)
    return label

#creating a button by one function
def createBtn(value, memory_name, job, width, row, column):
    button = QPushButton(value)
    button.setCursor(QCursor(QtCore.Qt.PointingHandCursor))
    button.setFixedSize(width, 50)
    button.setStyleSheet(
        "*{border: 2px solid '#2ED573';"+
        "color: #fff;"+
        "font-family: 'shanti';"+
        "font-size: 15px;"+
        "border-radius: 25px;"+
        "padding: 10px 0;}"+
        "*:hover{"+
        "background: '#2ED573';}"
    )
    widgets[memory_name].append(button)
    button.clicked.connect(job)
    grid.addWidget(button, row, column, alignment=Qt.AlignCenter)
    return button

#creating a back button by one function
def createBackButton(memory_name, job, row, column):
    #creating a normal button 
    back = createBtn("", memory_name, job, 50, row, column)
    #setting icon to the back button
    back.setIcon(QIcon("./imgs/back-icon.png"))

#creating a input feild by one function
def createInputFeild(memory_name, row, column):
    feild = QLineEdit()
    feild.setFixedSize(300, 50)
    feild.setStyleSheet(
        "*{border: 2px solid '#2ED573';"+
        "color: #fff;"+
        "font-family: 'shanti';"+
        "font-size: 13px;"+
        "border-radius: 25px;"+
        "padding: 15px 20px;}"
    )
    widgets[memory_name].append(feild)
    grid.addWidget(feild, row, column, alignment=Qt.AlignCenter)
    return feild

#creating an output layout 
def createOutput(out, row, column):
    #put the output into a label     
    output = QLabel("")
    output.setText(out)
    output.setStyleSheet(
        "color: #fff;"+
        "font-family: 'Rounded Mplus 1c';"+
        "font-size: 20px;"+
        "border: 2px solid '#2ED573'"
    )
    if widgets["output"] != []:
        widgets["output"][-1].hide()
        widgets["output"].pop()
    widgets["output"].append(output)
    grid.addWidget(output, row, column, alignment=Qt.AlignCenter)
    return output
#===========================================================================

#===============================METHODS====================================
#bisection method function
def bisection():
    #calling the clear function
    clearWidgets()

    #make a sub-program to calculate it
    def bisectionCalc():
        #if any input feild is empty
        if firstPiont.text() == "" or secondPoint.text() == "" or function.text() == "":
            #show the output     
            output = createOutput("Please Fill The Input Feilds", 3, 1)
            return
        else:
            #make a function of (x)
            def func(expr, x):
                #eval() to convert the string to integer
                return eval(expr)

            #getting the data from the input feilds
            input_expr = function.text()
            a = firstPiont.text()
            try:
                a = int(a)
            except ValueError:
                try:
                    a = float(a)
                except ValueError:
                    output = createOutput("Must Be a Number", 3, 1)
                    return
            b = secondPoint.text()
            try:
                b = int(b)
            except ValueError:
                try:
                    b = float(b)
                except ValueError:
                    output = createOutput("Must Be a Number", 3, 1)
                    return
            
            if (func(input_expr, a) * func(input_expr, b) >= 0): 
                output = createOutput("Not Allowed", 3, 1)
                return
        
            c = a 
            while ((b-a) >= 0.001): 
        
                # Find middle point 
                c = (a+b)/2
        
                # Check if middle point is root 
                if (func(input_expr,c) == 0.0): 
                    break
        
                # Decide the side to repeat the steps 
                if (func(input_expr,c)*func(input_expr,a) < 0): 
                    b = c 
                else: 
                    a = c 

            #show the output
            output = createOutput("Root = " + str(c), 3, 1)
            return

    #displaying logo
    logo = createLogo("./imgs/bisection.png")

    #creating inputs
    functionLabel = createLabel("Enter The Function of X : ", "label-one", 0, 0)
    function = createInputFeild("input-one", 1, 0)
    firstPointLabel = createLabel("Enter (a) Point : ", "label-two", 2, 0)
    firstPiont = createInputFeild("input-two", 3, 0)
    secondPointLabel = createLabel("Enter (b) Point", "label-three", 4, 0)
    secondPoint = createInputFeild("input-three", 5, 0)
    calc = createBtn("Calc", "button-one", bisectionCalc, 50, 6, 0)
    back = createBackButton("button-two", backToChoices, 6, 1)

#newton's method function
def newton():
    #calling the clear function
    clearWidgets()

    def newtonCalc():
        #if any input feild is empty
        if function.text() == "" or point.text() == "":
            #show the output       
            output = createOutput("Please Fill The Input Feilds", 2, 1)
            return
        else:
            #calculate the newton's method
            def NewtonsMethod(f, x, tolerance=0.001):
                while True:
                    x1 = x - f(x) / misc.derivative(f, x) 
                    t = abs(x1 - x)
                    if t < tolerance:
                        break
                    x = x1
                return x

            #make a function of x
            def f(x):
                return eval(function.text())
            
            #getting the data from the input feild
            x = point.text()
            try:
                x = int(x)
            except ValueError:
                try:
                    x = float(x)
                except ValueError:
                    output = createOutput("Must Be a Number", 2, 1)
                    return
            x0 = NewtonsMethod(f, x)

            #show the output       
            output = createOutput("Root = " + str(x0), 2, 1)
            return

    #display logo
    logo = createLogo("./imgs/newton.png")

    #create inputs
    functionLabel = createLabel("Enter The Function of X : ", "label-one", 0, 0)
    function = createInputFeild("input-one", 1, 0)
    pointLabel = createLabel("Enter (x) Point : ", "label-two", 2, 0)
    point = createInputFeild("input-two", 3, 0)
    calc = createBtn("Calc", "button-one", newtonCalc, 50, 4, 0)
    back = createBackButton("button-two", backToChoices, 4, 1)

#newton's jacobian method function
def trapazodial():
    #calling the clear function
    clearWidgets()

    def trapazodialCalc():
        #if any input feild is empty
        if funcInput.text() == "" or aInput.text() == "" or bInput.text() == "" or nInput.text() == "":
            #show the output       
            output = createOutput("Please Fill The Input Feilds", 4, 1)
            return
        else:
            def y( x ):
                # Declaring the function
                return eval(funcInput.text())
                
            # Function to evalute the value of integral
            def trapezoidal (a, b, n):
                try:
                    # Grid spacing
                    h = (b - a) / n
                    
                    # Computing sum of first and last terms
                    # in above formula
                    s = (y(a) + y(b))

                    # Adding middle terms in above formula
                    i = 1
                    while i < n:
                        s += 2 * y(a + i * h)
                        i += 1
                    # h/2 indicates (b-a)/2n.
                    # Multiplying h/2 with s.
                    return "Value of Integral is " + str(((h / 2) * s))
                except ZeroDivisionError:
                    return "Can't Divide By Zero"
                
            # Driver code to test above function
            # Range of definite integral
            xn = bInput.text()
            try:
                xn = int(xn)
            except ValueError:
                try:
                    xn = float(xn)
                except ValueError:
                    output = createOutput("Must Be a Number", 4, 1)
                    return
            x0 = aInput.text()
            try:
                x0 = int(x0)
            except ValueError:
                try:
                    x0 = float(x0)
                except ValueError:
                    output = createOutput("Must Be a Number", 4, 1)
                    return 

            # Number of grids. Higher value means
            n = nInput.text()
            try:
                n = int(n)
            except ValueError:
                try:
                    n = float(n)
                except ValueError:
                    output = createOutput("Must Be a Number", 4, 1)
                    return
            output = createOutput(trapezoidal(x0, xn, n), 4, 1)
            return

    logo = createLogo("./imgs/trap.png")

    #creating input fields
    funcLabel = createLabel("Enter The Function of X : ", "label-one", 0, 0)
    funcInput = createInputFeild("input-one", 1, 0)
    aLabel = createLabel("Enter The Initial Value : ", "label-two", 2, 0)
    aInput = createInputFeild("input-two", 3, 0)
    bLabel = createLabel("Enter The End Value : ", "label-three", 4, 0)
    bInput = createInputFeild("input-three", 5, 0)
    nLabel = createLabel("Enter The Number of Grids : ", "label-four", 6, 0)
    nInput = createInputFeild("input-four", 7, 0)
    calc = createBtn("Calc", "button-one", trapazodialCalc, 50, 8, 0)
    back = createBackButton("button-two", backToChoices, 8, 1)

#lagarange method function
def lagarange():
    #calling the clear function
    clearWidgets()

    def lagCalc():
        #if any input feild is empty
        if xInputs.text() == "" or yInputs.text() == "" or interpolation.text() == "":
            #show the output       
            output = createOutput("Please Fill The Input Feilds", 3, 1)
            return
        else:
            #getting inputs from the user
            x = eval(xInputs.text())
            y = eval(yInputs.text())
            
            #(len()) to know the number of points
            n = len(x)
            # Reading data points
            for i in range(n):
                x[i] = x[i]
                y[i] = y[i]


            # getting interpolation point from the user
            xp = interpolation.text()
            try:
                xp = int(xp)
            except ValueError:
                try:
                    xp = float(xp)
                except ValueError:
                    output = createOutput("Must Be a Number", 3, 1)
                    return
            # Set interpolated value initially to zero
            yp = 0

            # Implementing Lagrange Interpolation
            for i in range(n):
                
                p = 1
                
                for j in range(n):
                    if i != j:
                        p = p * (xp - x[j])/(x[i] - x[j])
                
                yp = yp + p * y[i]    

                #show the output       
                output = createOutput("Interpolated value is " + str(yp), 3, 1)
            return
    
    #display logo
    logo = createLogo("./imgs/lagarange.png")

    #create inputs
    xInputsLabel = createLabel("Enter [x] items in This Form [-, -, -, -] : ", "label-one", 0, 0)
    xInputs = createInputFeild("input-one", 1, 0)
    yInputsLabel = createLabel("Enter [y] items in This Form [-, -, -, -] : ", "label-two", 2, 0)
    yInputs = createInputFeild("input-two", 3, 0)
    interpolationLabel = createLabel("Enter The Interpolation : ", "label-three", 4, 0)
    interpolation = createInputFeild("input-three", 5, 0)
    calc = createBtn("Calc", "button-one", lagCalc, 50, 6, 0)
    back = createBackButton("button-two", backToChoices, 6, 1)

#devided difference method function
def devided():
    #calling the clear function
    clearWidgets()
    
    def dividedCalc():
        if xInput.text() == "" or yInput.text() == "" or valueInput.text() == "":
            #show the output       
            output = createOutput("Please Fill The Input Feilds", 3, 1)
            return
        else:
            # Python3 program for implementing 
            # Newton divided difference formula 
            
            # Function to find the product term 
            def proterm(i, value, x): 
                pro = 1 
                for j in range(i): 
                    pro = pro * (value - x[j])
                return pro 
            
            # Function for calculating 
            # divided difference table 
            def dividedDiffTable(x, y, n):
            
                for i in range(1, n): 
                    for j in range(n - i): 
                        y[j][i] = ((y[j][i - 1] - y[j + 1][i - 1]) /
                                                (x[j] - x[i + j]))
                return y
            
            # Function for applying Newton's 
            # divided difference formula 
            def applyFormula(value, x, y, n): 
            
                sum = y[0][0] 
            
                for i in range(1, n):
                    sum = sum + (proterm(i, value, x) * y[0][i]) 
                
                return sum 
            
            # Function for displaying divided 
            # difference table 
            def printDiffTable(y, n):
                for i in range(n): 
                    for j in range(n - i): 
                        output = createOutput(eval(round(y[i][j],6), "   "), 0, 1)
            
            # Driver Code
            # number of inputs given 
            n = len(eval(xInput.text()))

            y = [[0 for i in range(10)] 
                    for j in range(10)]

            # number of x
            x = eval(xInput.text())


            # y[][] is used for divided difference 
            # table where y[][0] is used for input 
            yi = eval(yInput.text())

            for i in range(n):
                y[i][0] = yi[i]

            # calculating divided difference table 
            y=dividedDiffTable(x, y, n) 
            
            # displaying divided difference table 
            
            # value to be interpolated 
            value = valueInput.text()
            try:
                value = int(value)
            except ValueError:
                try:
                    value = float(value)
                except ValueError:
                    output = createOutput("Must Be a Number", 3, 1)
                    return
            output = createOutput("Value is " + str(round(applyFormula(value, x, y, n), 2)), 3, 1)
            return

    logo = createLogo("./imgs/divided.png")

    xInputLabel = createLabel("Enter [x] items in This Form [-, -, -, -] : ", "label-one", 0, 0)
    xInput = createInputFeild("input-one", 1, 0)
    yInputLabel = createLabel("Enter [y] items in This Form [-, -, -, -] : ", "label-two", 2, 0)
    yInput = createInputFeild("input-two", 3, 0)
    valueLabel = createLabel("Enter The Value : ", "label-three", 4, 0)
    valueInput = createInputFeild("input-three", 5, 0)
    calc = createBtn("Calc", "button-one", dividedCalc, 50, 6, 0)
    back = createBackButton("button-two", backToChoices, 6, 1)
#==========================================================================

#===============================MAIN FRAMES====================================
#starting frame
def startingFrame():
    #display logo
    logo = createLogo("./imgs/logo.png")

    #create, display and styling the start button
    startButton = createBtn("Start", "button-one", start, 200, 1, 2)

#choices of the methods frame
def choicesFrame():
    #display logo
    logo = createLogo("./imgs/logo.png")

    label = createLabel("Choose The Method You Want : ", "label-one", 0, 0)
    
    #methods buttons
    bisecMethodBtn = createBtn("Bisection Method", "button-one", bisection, 200, 1, 0)
    newMethodBtn = createBtn("Newton's Method", "button-two", newton, 200, 1, 1)
    trapMethodBtn = createBtn("Trapazodial Method", "button-three", trapazodial, 200, 1, 2)
    lagMethodBtn = createBtn("Lagrange Method", "button-four", lagarange, 200, 2, 0)
    devMethodBtn = createBtn("Divided differences Method", "button-fife", devided, 200, 2, 1)
    back = createBackButton("button-six", backToStart, 2, 2)
#============================================================================

#setting a layout to display the widgets on the window
window.setLayout(mainLayout)
mainLayout.addLayout(grid, 1, 0)

#calling the 1st fram or the starting frame to be displayed while the program starting
startingFrame()

#displaying the window from the memory
window.show()
sys.exit(app.exec())
#===============================END OF THE CODE====================================
```
